# **Ejercicio 1**
Una revision de los pesos de 500 estudiantes del ITD, evidenci√≥ que la media es de 70 kilogramos y la desviacion estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:

Generar aleatoriamente las edades de los 500 estudiantes con una media de 70 kg y una desviacion estandar de 4kg.

Realizar un historama de las edades.

Graficar la funcion de densidad de probabilidad(PDF)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#Generar datos para la distribucion normal

media = 70 #Media
desv_std = 4 #Desviacion estandar
num_datos =500

#Generamos los datos de manera aleatoria y los ponemos en una lista (array)
datos= np.random.normal(media, desv_std, num_datos)
datos[0:50] #Mostrar los primeros 50
```

```python
#Crear un rango de valores para crear una densidad
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)

#Calculas la pdf de la distribucion normal
pdf=1/ desv_std * np.sqrt(2 * np.pi) * np.exp(-(x - media) **2 / (2 * desv_std **2))

```

```python
#Para no imprimir toda la lista de valores (500), desplegamos los primeros 50

print("Imprimir los primero 50 valores de x",x[0:50],", Y los primero 50 valores de pdf",pdf[0:50])
```

```python
#Graficar historama
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')

plt.title('Distribucion normal')
plt.xlabel('Peso (kg)De los estudiantes')
plt.grid(True)
plt.show()

```

```python
# visualizar la PDF
plt.figure(figsize=(6,4))
plt.plot(x, pdf, label='normal:(media:{};ds:{})'.format(np.round(media,2), np.round(desv_std,2)))
plt.title('Funcion de densidad de probabilidad (PDF)- Distribucion normal')
plt.xlabel('Peso (kg)De los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python

#Graficar Histograma
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')

plt.title('Distribucion normal')
plt.xlabel('Peso (kg)De los estudiantes')
plt.grid(True)
plt.show()
```

# **Ejercicio 2**
Una revision de los pesos de 500 estudiantes del ITD, evidencio que la media es de 70 kilogramos y la desviacion estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente.

Cuantos estudiantes pesan menos de 75 kg.
Cuantos estudiantes oesan menos de 62 kg.
Cuantos estudiantes pesan mas de 79 kg.

```python

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
plt.figure(figsize=(6,4))
#Parametros de distribucion normal
media = 70
desv_std = 4 #Desviaicon estandar
total_estudiantes = 500

#Variable para el calculo de la probabilidad de que un estudiante pese menos de 75kg
peso_limite = 79

#Calcular la probabilidad de que un estudiante pese menos
# de 75 kg utilizando la funcion de distribucion acumulativa (CDF) de la distribucion normal
prob_menos_75kg=norm.cdf(peso_limite, media, desv_std )
```

```python
#calcular el numero de estudiantes
num_estudiantes_75kg = prob_menos_75kg * total_estudiantes

print("Probabilidad de que los estudiantes pesen menos de 75kg", prob_menos_75kg)
print("Numero de estudiantes que pesan menos de 75kg", num_estudiantes_75kg)


# graficar
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 100)
y = stats.norm.pdf(x, media, desv_std)

plt.plot(x, y, label='Distribuci√≥n Normal')
plt.axvline(peso_limite, color='red', linestyle='--', label='Peso Limite')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribuci√≥n Normal de Pesos de Estudiantes')
plt.legend()
plt.grid(True)
plt.show()
```

```python

#Cuanto estudiantes pesan mendo de 62kg
plt.figure(figsize=(6,4))

#Parametros de la distribucion normal
media = 70
desv_std = 4
total_estudiantes = 500

#Variable para calculo de la probabilidad que un estudiante pese menos de 75kg
peso_limite = 62

#Necesitamos calculas la probabilidad de que un estudiante pese menos de 75kg utilizando la funcion de distribucion
# acumulativa (CDF) de la distribucion normal
prob_menos_62kg = stats.norm.cdf(peso_limite, media, desv_std)

#Calculo del numero de estudiantes que pesan menos de  75 kg
estudiantes_menos_62kg = prob_menos_62kg * total_estudiantes

print("Probabilidad de que los estudiantes pesen menos de 62kg", prob_menos_62kg)
print("Numero de estudiantes que pesan menos de 75kg", estudiantes_menos_62kg)
```

```python
#Calcular la distribucion normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = stats.norm.pdf(x, media, desv_std)

#Graficar la distribucion normal
plt.plot(x, y, "b", linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color="skyblue", alpha=0.5)
plt.title("Probabilidad de que un estudiante pese menos de 62 kg,")
plt.xlabel("Peso (kg)")
plt.ylabel("Densidad de Probabilidad")
plt.grid(True)
plt.show()

```
